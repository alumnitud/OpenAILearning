```python
import os
import sys
import datetime
import openai

from dotenv import load_dotenv
```

## Function

Loading env variables from .env file

```python
dotenv_path = ".env"
load_dotenv(dotenv_path=dotenv_path)

client = openai.AzureOpenAI(
        azure_endpoint= os.getenv("AZURE_API_ENDPOINT"),
        api_key= os.getenv("AZURE_API_KEY"),
        api_version= os.getenv("AZURE_API_VERSION") 
)
```

Defining the function

```python
def get_response(prompt):
    conversation = [{"role": "system", "content": """You are a teacher asistant. You should generate some lesson plan with some numbered list.Agenda: """}]

    conversation.append({"role": "user", "content": prompt})

    completion = client.chat.completions.create(
        model="gpt-4-turbo",
        messages=conversation
    )

    answer = "\n" + completion.choices[0].message.content + "\n"
    return answer
```

## Testing

```python
prompt = "I want to study Machine Learning Foundations in 1 week"
print(get_response(prompt))
```
